```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("../data/cleaned_superstore.csv")

df.head()
```

```python
engine = create_engine("sqlite:///superstore.db")

df.to_sql(
    "superstore",
    engine,
    if_exists="replace",
    index=False
)

print("Database Created Successfully")
```

```python
query = """
SELECT *
FROM superstore
LIMIT 5;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT
    Category,
    Sales
FROM superstore
LIMIT 10;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT *
FROM superstore
WHERE Sales > 500;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT
    Product_Name,
    Sales
FROM superstore
ORDER BY Sales DESC
LIMIT 10;
"""
```

```python
query = """
SELECT
    [Product Name],
    SUM(Sales) AS TotalSales
FROM superstore
GROUP BY [Product Name]
ORDER BY TotalSales DESC
LIMIT 5;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT
    [Customer Name],
    SUM(Sales) AS TotalSales
FROM superstore
GROUP BY [Customer Name]
ORDER BY TotalSales DESC
LIMIT 5;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT
    Category,
    SUM(Sales) AS TotalSales
FROM superstore
GROUP BY Category;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT
    Region,
    SUM(Profit) AS TotalProfit
FROM superstore
GROUP BY Region;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT *
FROM superstore
WHERE Sales >
(
    SELECT AVG(Sales)
    FROM superstore
);
"""

pd.read_sql(query, engine)
```

```python
query = """
WITH CategorySales AS
(
    SELECT
        Category,
        SUM(Sales) AS TotalSales
    FROM superstore
    GROUP BY Category
)

SELECT *
FROM CategorySales;
"""

pd.read_sql(query, engine)
```

```python
query = """
SELECT
    [Customer Name],
    Sales,
    RANK() OVER
    (
        ORDER BY Sales DESC
    ) AS SalesRank
FROM superstore;
"""

pd.read_sql(query, engine)
```
